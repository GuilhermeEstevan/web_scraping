```python
import requests

# URL do PDF
url = "https://www.gov.br/receitafederal/dados/cnpj-metadados.pdf"
output_path = "cnpj-metadados.pdf"

# Fazer a requisição para baixar o PDF
response = requests.get(url, stream=True)
if response.status_code == 200:
    with open(output_path, "wb") as f:
        f.write(response.content)
    print(f"PDF baixado com sucesso em {output_path}")
else:
    print(f"Falha ao baixar o PDF. Código de status: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import os

# URL da página que contém os links para os arquivos ZIP
base_url = "https://arquivos.receitafederal.gov.br/cnpj/dados_abertos_cnpj/2024-10/"
output_folder = "dados_cnpj"

# Criar a pasta para salvar os arquivos ZIP, se ela ainda não existir
os.makedirs(output_folder, exist_ok=True)

# Fazer a requisição para obter o HTML da página
response = requests.get(base_url)
soup = BeautifulSoup(response.content, "html.parser")

# Encontrar todos os links para arquivos ZIP
links = soup.find_all("a")
zip_links = [base_url + link.get("href") for link in links if link.get("href") and link.get("href").endswith(".zip")]

# Baixar cada arquivo ZIP
for zip_link in zip_links:
    file_name = zip_link.split("/")[-1]  # Nome do arquivo a partir da URL
    file_path = os.path.join(output_folder, file_name)
    
    print(f"Baixando {file_name}...")
    response = requests.get(zip_link, stream=True)
    
    if response.status_code == 200:
        with open(file_path, "wb") as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        print(f"{file_name} baixado com sucesso!")
    else:
        print(f"Falha ao baixar {file_name}. Código de status: {response.status_code}")
```

```python
import zipfile
import os

# Caminho para a pasta onde os arquivos ZIP foram salvos
zip_folder = "dados_cnpj"
extracted_folder = "dados_cnpj_extraidos"

# Criar a pasta para salvar os arquivos extraídos, se ela ainda não existir
os.makedirs(extracted_folder, exist_ok=True)

# Extrair todos os arquivos ZIP
for zip_file in os.listdir(zip_folder):
    if zip_file.endswith(".zip"):
        zip_path = os.path.join(zip_folder, zip_file)
        with zipfile.ZipFile(zip_path, "r") as zip_ref:
            zip_ref.extractall(extracted_folder)
            print(f"{zip_file} extraído com sucesso!")
```

```python
import pandas as pd
import psycopg2
from psycopg2 import extras

def inserir_empresas_em_chunks(conn, csv_path, chunksize=10000, batch_size=1000):
    try:
        cursor = conn.cursor()
        total_linhas_inseridas = 0

        for chunk in pd.read_csv(csv_path, delimiter=';', encoding='latin1', header=None, chunksize=chunksize):
            chunk.columns = [
                "cnpj_basico", "razao_social", "natureza_juridica", 
                "qualificacao_responsavel", "capital_social", 
                "porte_empresa", "ente_federativo_responsavel"
            ]

            # Tratando valores "NaN" e convertendo tipos de forma segura
            chunk['capital_social'] = chunk['capital_social'].replace({',': ''}, regex=True).astype(float)
            chunk['porte_empresa'] = pd.to_numeric(chunk['porte_empresa'], errors='coerce').fillna(0).astype(int)
            chunk['ente_federativo_responsavel'] = chunk['ente_federativo_responsavel'].fillna('')

            # Dividindo o chunk em lotes menores
            for i in range(0, len(chunk), batch_size):
                batch = chunk.iloc[i:i+batch_size]
                values = [tuple(row) for row in batch.values]
                
                insert_query = """
                INSERT INTO empresas (cnpj_basico, razao_social, natureza_juridica, 
                                      qualificacao_responsavel, capital_social, porte_empresa, 
                                      ente_federativo_responsavel)
                VALUES %s
                """
                extras.execute_values(cursor, insert_query, values)
                total_linhas_inseridas += len(values)

            conn.commit()  # Faz o commit após cada chunk
            print(f"{total_linhas_inseridas} linhas inseridas até agora.")

        cursor.close()
        print("Dados inseridos na tabela empresas com sucesso.")
    except Exception as e:
        print(f"Erro ao inserir dados na tabela empresas: {e}")

```

```python
def inserir_empresas_com_qualificacoes_automaticas(conn, csv_path, chunksize=10000, batch_size=1000):
    try:
        cursor = conn.cursor()
        total_linhas_inseridas = 0

        for chunk in pd.read_csv(csv_path, delimiter=';', encoding='latin1', header=None, chunksize=chunksize):
            chunk.columns = [
                "cnpj_basico", "razao_social", "natureza_juridica", 
                "qualificacao_responsavel", "capital_social", 
                "porte_empresa", "ente_federativo_responsavel"
            ]

            # Tratando valores "NaN" e convertendo tipos de forma segura
            chunk['capital_social'] = chunk['capital_social'].replace({',': ''}, regex=True).astype(float)
            chunk['porte_empresa'] = pd.to_numeric(chunk['porte_empresa'], errors='coerce').fillna(0).astype(int)
            chunk['ente_federativo_responsavel'] = chunk['ente_federativo_responsavel'].fillna('')

            # Converte as colunas para tipos Python nativos
            chunk = chunk.astype({
                "cnpj_basico": str,
                "razao_social": str,
                "natureza_juridica": int,
                "qualificacao_responsavel": int,
                "capital_social": float,
                "porte_empresa": int,
                "ente_federativo_responsavel": str
            })

            # Verificar e adicionar qualificacoes faltantes
            qualificacoes_unicas = chunk['qualificacao_responsavel'].dropna().unique()
            cursor.execute("SELECT codigo FROM qualificacoes")
            qualificacoes_existentes = {row[0] for row in cursor.fetchall()}
            
            qualificacoes_para_inserir = [
                (int(qualificacao), f"Qualificação {qualificacao}") 
                for qualificacao in qualificacoes_unicas if qualificacao not in qualificacoes_existentes
            ]
            
            if qualificacoes_para_inserir:
                cursor.executemany(
                    "INSERT INTO qualificacoes (codigo, descricao) VALUES (%s, %s) ON CONFLICT DO NOTHING", 
                    qualificacoes_para_inserir
                )
                conn.commit()
                print(f"{len(qualificacoes_para_inserir)} novas qualificações inseridas.")

            # Dividir o chunk em lotes menores para inserções em lote
            for i in range(0, len(chunk), batch_size):
                batch = chunk.iloc[i:i+batch_size]
                # Converte o DataFrame para uma lista de tuplas com tipos nativos do Python
                values = [tuple(row) for row in batch.values]
                
                insert_query = """
                INSERT INTO empresas (cnpj_basico, razao_social, natureza_juridica, 
                                      qualificacao_responsavel, capital_social, porte_empresa, 
                                      ente_federativo_responsavel)
                VALUES %s
                """
                extras.execute_values(cursor, insert_query, values)
                total_linhas_inseridas += len(values)

            conn.commit()  # Faz o commit após cada chunk
            print(f"{total_linhas_inseridas} linhas inseridas até agora.")

        cursor.close()
        print("Dados inseridos na tabela empresas com sucesso.")
    except Exception as e:
        print(f"Erro ao inserir dados na tabela empresas: {e}")
```

```python
import pandas as pd
import psycopg2
from psycopg2 import extras

def inserir_socios_em_chunks(conn, csv_path, chunksize=10000, batch_size=1000):
    try:
        cursor = conn.cursor()
        total_linhas_inseridas = 0

        for chunk in pd.read_csv(csv_path, delimiter=';', encoding='latin1', header=None, chunksize=chunksize):
            chunk.columns = [
                "cnpj_basico", "identificador_socio", "nome_socio", 
                "cpf_cnpj_socio", "codigo_qualificacao", 
                "data_entrada", "pais", "representante_legal", 
                "coluna_descartada",  # Nome do Representante, que será descartado
                "qualificacao_representante", "faixa_etaria"
            ]

            # Remover a coluna "Nome do Representante" que está sendo descartada
            chunk = chunk.drop(columns=["coluna_descartada"])

            # Tratando valores "NaN" e convertendo tipos de forma segura
            chunk['data_entrada'] = pd.to_datetime(chunk['data_entrada'], errors='coerce').dt.date
            chunk['pais'] = chunk['pais'].fillna(0).astype(int)

            # Converte colunas para tipos Python nativos
            chunk = chunk.astype({
                "cnpj_basico": str,
                "identificador_socio": int,
                "nome_socio": str,
                "cpf_cnpj_socio": str,
                "codigo_qualificacao": int,
                "pais": int,
                "representante_legal": str,
                "qualificacao_representante": int,
                "faixa_etaria": int
            })

            # Verificar e adicionar qualificações faltantes
            qualificacoes_unicas = chunk['codigo_qualificacao'].dropna().unique()
            cursor.execute("SELECT codigo FROM qualificacoes")
            qualificacoes_existentes = {row[0] for row in cursor.fetchall()}
            
            qualificacoes_para_inserir = [
                (int(qualificacao), f"Qualificação {qualificacao}") 
                for qualificacao in qualificacoes_unicas if qualificacao not in qualificacoes_existentes
            ]
            
            if qualificacoes_para_inserir:
                cursor.executemany(
                    "INSERT INTO qualificacoes (codigo, descricao) VALUES (%s, %s) ON CONFLICT DO NOTHING", 
                    qualificacoes_para_inserir
                )
                conn.commit()
                print(f"{len(qualificacoes_para_inserir)} novas qualificações inseridas.")

            # Verificar e adicionar países faltantes
            paises_unicos = chunk['pais'].dropna().unique()
            cursor.execute("SELECT codigo FROM paises")
            paises_existentes = {row[0] for row in cursor.fetchall()}
            
            paises_para_inserir = [
                (int(pais), f"País {pais}") 
                for pais in paises_unicos if pais not in paises_existentes and pais != 0
            ]
            
            if paises_para_inserir:
                cursor.executemany(
                    "INSERT INTO paises (codigo, nome) VALUES (%s, %s) ON CONFLICT DO NOTHING", 
                    paises_para_inserir
                )
                conn.commit()
                print(f"{len(paises_para_inserir)} novos países inseridos.")

            # Dividir o chunk em lotes menores para inserções em lote
            for i in range(0, len(chunk), batch_size):
                batch = chunk.iloc[i:i+batch_size]
                values = [tuple(row) for row in batch.values]
                
                insert_query = """
                INSERT INTO socios (cnpj_basico, identificador_socio, nome_socio, 
                                    cpf_cnpj_socio, codigo_qualificacao, data_entrada, 
                                    pais, representante_legal, 
                                    qualificacao_representante, faixa_etaria)
                VALUES %s
                ON CONFLICT (cnpj_basico, identificador_socio) DO NOTHING
                """
                extras.execute_values(cursor, insert_query, values)
                total_linhas_inseridas += len(values)

            conn.commit()  # Faz o commit após cada chunk
            print(f"{total_linhas_inseridas} linhas inseridas até agora.")

        cursor.close()
        print("Dados inseridos na tabela socios com sucesso.")
    except Exception as e:
        print(f"Erro ao inserir dados na tabela socios: {e}")
```

```python
import pandas as pd
import psycopg2
from psycopg2 import extras

def inserir_dados_simples_em_chunks(conn, csv_path, chunksize=10000, batch_size=1000):
    try:
        cursor = conn.cursor()
        total_linhas_inseridas = 0

        for chunk in pd.read_csv(csv_path, delimiter=';', encoding='latin1', header=None, chunksize=chunksize):
            # Nomeando as colunas com base na estrutura da tabela
            chunk.columns = [
                "cnpj_basico", "op_simples_nacional", "data_opcao_simples",
                "data_exclusao_simples", "op_mei", "data_opcao_mei", "data_exclusao_mei"
            ]

            # Tratando as colunas de data, substituindo "00000000" por None e formatando as datas
            for col in ["data_opcao_simples", "data_exclusao_simples", "data_opcao_mei", "data_exclusao_mei"]:
                chunk[col] = pd.to_datetime(chunk[col].replace("00000000", None), format='%Y%m%d', errors='coerce')

            # Convertendo `NaT` para `None`
            chunk = chunk.where(pd.notnull(chunk), None)

            # Converte colunas para tipos Python nativos
            chunk = chunk.astype({
                "cnpj_basico": str,
                "op_simples_nacional": str,
                "op_mei": str
            })

            # Dividindo o chunk em lotes menores para inserções em lote
            for i in range(0, len(chunk), batch_size):
                batch = chunk.iloc[i:i+batch_size]
                
                # Prepara os valores, convertendo `NaT` em `None`
                values = [(
                    row['cnpj_basico'], row['op_simples_nacional'],
                    row['data_opcao_simples'] if pd.notnull(row['data_opcao_simples']) else None,
                    row['data_exclusao_simples'] if pd.notnull(row['data_exclusao_simples']) else None,
                    row['op_mei'], 
                    row['data_opcao_mei'] if pd.notnull(row['data_opcao_mei']) else None,
                    row['data_exclusao_mei'] if pd.notnull(row['data_exclusao_mei']) else None
                ) for _, row in batch.iterrows()]

                # Definindo a query de inserção
                insert_query = """
                INSERT INTO dados_simples (cnpj_basico, op_simples_nacional, data_opcao_simples, 
                                           data_exclusao_simples, op_mei, data_opcao_mei, data_exclusao_mei)
                VALUES %s
                """
                extras.execute_values(cursor, insert_query, values)
                total_linhas_inseridas += len(values)

            conn.commit()  # Faz o commit após cada chunk
            print(f"{total_linhas_inseridas} linhas inseridas até agora.")

        cursor.close()
        print("Dados inseridos na tabela dados_simples com sucesso.")
    except Exception as e:
        print(f"Erro ao inserir dados na tabela dados_simples: {e}")

```

```python
import pandas as pd
import psycopg2
from psycopg2 import extras

def get_total_rows(conn):
    """Retorna o número total de linhas já inseridas na tabela."""
    cursor = conn.cursor()
    cursor.execute("SELECT COUNT(*) FROM dados_simples")
    total_rows = cursor.fetchone()[0]
    cursor.close()
    return total_rows

def inserir_dados_simples_em_chunks(conn, csv_path, chunksize=10000, batch_size=1000, skip_rows=0):
    try:
        cursor = conn.cursor()
        total_linhas_inseridas = skip_rows

        for chunk in pd.read_csv(csv_path, delimiter=';', encoding='latin1', header=None, 
                                 chunksize=chunksize, skiprows=skip_rows + 1):  # +1 para ignorar o cabeçalho
            # Nomeando as colunas com base na estrutura da tabela
            chunk.columns = [
                "cnpj_basico", "op_simples_nacional", "data_opcao_simples",
                "data_exclusao_simples", "op_mei", "data_opcao_mei", "data_exclusao_mei"
            ]

            # Tratando as colunas de data, substituindo "00000000" por None e formatando as datas
            for col in ["data_opcao_simples", "data_exclusao_simples", "data_opcao_mei", "data_exclusao_mei"]:
                chunk[col] = pd.to_datetime(chunk[col].replace("00000000", None), format='%Y%m%d', errors='coerce')

            # Convertendo `NaT` para `None`
            chunk = chunk.where(pd.notnull(chunk), None)

            # Converte colunas para tipos Python nativos
            chunk = chunk.astype({
                "cnpj_basico": str,
                "op_simples_nacional": str,
                "op_mei": str
            })

            # Dividindo o chunk em lotes menores para inserções em lote
            for i in range(0, len(chunk), batch_size):
                batch = chunk.iloc[i:i+batch_size]
                
                # Prepara os valores, convertendo `NaT` em `None`
                values = [(
                    row['cnpj_basico'], row['op_simples_nacional'],
                    row['data_opcao_simples'] if pd.notnull(row['data_opcao_simples']) else None,
                    row['data_exclusao_simples'] if pd.notnull(row['data_exclusao_simples']) else None,
                    row['op_mei'], 
                    row['data_opcao_mei'] if pd.notnull(row['data_opcao_mei']) else None,
                    row['data_exclusao_mei'] if pd.notnull(row['data_exclusao_mei']) else None
                ) for _, row in batch.iterrows()]

                # Definindo a query de inserção
                insert_query = """
                INSERT INTO dados_simples (cnpj_basico, op_simples_nacional, data_opcao_simples, 
                                           data_exclusao_simples, op_mei, data_opcao_mei, data_exclusao_mei)
                VALUES %s
                """
                extras.execute_values(cursor, insert_query, values)
                total_linhas_inseridas += len(values)

            conn.commit()  # Faz o commit após cada chunk
            print(f"{total_linhas_inseridas} linhas inseridas até agora.")

        cursor.close()
        print("Dados inseridos na tabela dados_simples com sucesso.")
    except Exception as e:
        print(f"Erro ao inserir dados na tabela dados_simples: {e}")

```

```python
import pandas as pd

def checar_valores_acima_do_limite(csv_path, chunksize=10000):
    # Definindo os limites de acordo com os tipos das colunas no banco
    limites_colunas = {
        "identificador_matriz_filial": (0, 2147483647),   # int4
        "situacao_cadastral": (0, 2147483647),            # int4
        "motivo_situacao_cadastral": (0, 2147483647),     # int4
        "pais": (0, 2147483647),                          # int4
        "cnae_fiscal_principal": (0, 2147483647),         # int4
        "municipio": (0, 2147483647)                      # int4
    }

    # Dicionário para armazenar colunas e valores acima do limite
    valores_acima_do_limite = {col: [] for col in limites_colunas.keys()}

    # Lendo o arquivo CSV em chunks
    for chunk in pd.read_csv(csv_path, delimiter=';', encoding='latin1', header=None, chunksize=chunksize):
        # Nomeando as colunas de acordo com a tabela
        chunk.columns = [
            "cnpj_basico", "cnpj_ordem", "cnpj_dv", "identificador_matriz_filial",
            "nome_fantasia", "situacao_cadastral", "data_situacao_cadastral",
            "motivo_situacao_cadastral", "cidade_exterior", "pais", "data_inicio_atividade",
            "cnae_fiscal_principal", "cnae_fiscal_secundario", "tipo_logradouro", "logradouro",
            "numero", "complemento", "bairro", "cep", "uf", "municipio", "ddd1", "telefone1",
            "ddd2", "telefone2", "ddd_fax", "fax", "email", "situacao_especial", "data_situacao_especial"
        ]

        # Convertendo as colunas relevantes para numérico, se possível, e verificando os limites
        for col, (min_val, max_val) in limites_colunas.items():
            chunk[col] = pd.to_numeric(chunk[col], errors='coerce')

            # Identificando valores fora do intervalo permitido
            valores_fora_do_limite = chunk[(chunk[col] < min_val) | (chunk[col] > max_val)][col]
            if not valores_fora_do_limite.empty:
                valores_acima_do_limite[col].extend(valores_fora_do_limite.tolist())

    # Exibindo os resultados
    for col, valores in valores_acima_do_limite.items():
        if valores:
            print(f"Coluna '{col}' possui valores fora do limite permitido ({limites_colunas[col]}):")
            print(valores)
        else:
            print(f"Coluna '{col}' está dentro do limite permitido.")

# Caminho para o arquivo CSV
csv_path = "/home/guilherme/web_scrapy/dados_cnpj_extraidos/K3241.K03200Y0.D41012.ESTABELE"

# Chamando a função para verificar valores acima do limite
checar_valores_acima_do_limite(csv_path)
```

```python
# Função para truncar e reinserir dados no formato correto
def truncar_e_inserir_municipios(conn, csv_path, table_name, column_names):
    try:
        # Truncar a tabela para limpar dados antigos
        cursor = conn.cursor()
        cursor.execute(f"TRUNCATE TABLE {table_name} RESTART IDENTITY CASCADE;")
        conn.commit()
        
        print(f"Tabela {table_name} truncada com sucesso.")

        # Carregar o CSV com pandas
        df = pd.read_csv(csv_path, delimiter=';', encoding='latin1', header=None, dtype=str)
        df.columns = column_names

        # Garantir que o código dos municípios tenha sempre 4 dígitos (adicionando zero à esquerda quando necessário)
        df['codigo'] = df['codigo'].str.zfill(4)

        # Inserir os dados na tabela
        for _, row in df.iterrows():
            values = tuple(row)
            insert_query = f"INSERT INTO {table_name} ({', '.join(column_names)}) VALUES ({', '.join(['%s'] * len(column_names))})"
            cursor.execute(insert_query, values)

        conn.commit()
        cursor.close()
        print(f"Dados inseridos na tabela {table_name} com sucesso.")
    except Exception as e:
        print(f"Erro ao inserir dados na tabela {table_name}: {e}")

# Conectar ao banco e inserir os dados
conn = conectar_banco()
truncar_e_inserir_municipios(
    conn,
    csv_path="/home/guilherme/web_scrapy/dados_cnpj_extraidos/F.K03200$Z.D41012.MUNICCSV",
    table_name="municipios",
    column_names=["codigo", "nome"]
)
conn.close()
```

```python
import pandas as pd
import psycopg2
from psycopg2 import extras
from datetime import datetime

# Função para converter datas no formato AAAA-MM-DD para datetime ou None
def parse_date(date_str):
    if pd.isna(date_str) or date_str in ["", "NaT", "NaN"]:
        return None
    try:
        return datetime.strptime(date_str, "%Y-%m-%d")
    except ValueError:
        return None

# Função para limpar dados, convertendo '0' em campos específicos para None
def limpar_dados(row):
    for campo in ["ddd1", "telefone1", "ddd2", "telefone2", "ddd_fax", "fax", "municipio"]:
        if row[campo] == "0":
            row[campo] = None
    return row

# Função para verificar se um valor está dentro do intervalo de int4 ou int2
def verificar_limite_inteiro(valor, limite="int4"):
    try:
        if valor is None:
            return True
        valor = int(valor)
        if limite == "int4" and -2147483648 <= valor <= 2147483647:
            return True
        elif limite == "int2" and -32768 <= valor <= 32767:
            return True
        else:
            return False
    except ValueError:
        return False

# Função para inserir dados do arquivo corrigido em chunks
def inserir_dados_em_chunks(conn, csv_path, chunksize=10000):
    try:
        total_linhas_inseridas = 0
        cursor = conn.cursor()
        
        # Processamento em chunks
        for chunk in pd.read_csv(csv_path, delimiter=';', encoding='latin1', dtype={
            "cnpj_ordem": str, "cnpj_dv": str, "identificador_matriz_filial": str,
            "situacao_cadastral": str, "motivo_situacao_cadastral": str, 
            "municipio": str, "ddd1": str, "telefone1": str, "ddd2": str, 
            "telefone2": str, "ddd_fax": str, "fax": str, "complemento": str
        }, chunksize=chunksize):

            # Limpeza de dados específicos para transformar '0' em None onde necessário
            chunk = chunk.apply(limpar_dados, axis=1)

            # Garantir que "NaN" seja convertido para None
            chunk = chunk.map(lambda x: None if pd.isna(x) or x == "NaN" else x)

            # Aplicar a função de conversão de datas nas colunas de data
            for col in ["data_situacao_cadastral", "data_inicio_atividade", "data_situacao_especial"]:
                chunk[col] = chunk[col].apply(parse_date)

            # Preparar os dados para inserção
            values = []
            for _, row in chunk.iterrows():
                # Verificar limites para campos numéricos específicos
                if not verificar_limite_inteiro(row['municipio'], "int4"):
                    print(f"Erro de limite no campo 'municipio': {row['municipio']} - Email: {row['email']}")
                    continue
                if not verificar_limite_inteiro(row['cnae_fiscal_principal'], "int4"):
                    print(f"Erro de limite no campo 'cnae_fiscal_principal': {row['cnae_fiscal_principal']} - Email: {row['email']}")
                    continue
                
                # Prepara uma linha para inserção
                value = (
                    row['cnpj_basico'], row['cnpj_ordem'], row['cnpj_dv'],
                    row['identificador_matriz_filial'], row['nome_fantasia'],
                    row['situacao_cadastral'], row['data_situacao_cadastral'],
                    row['motivo_situacao_cadastral'] if row['motivo_situacao_cadastral'] else None,
                    row['cidade_exterior'], row['pais'], row['data_inicio_atividade'],
                    row['cnae_fiscal_principal'], row['cnae_fiscal_secundario'], row['tipo_logradouro'], 
                    row['logradouro'], row['numero'], row['complemento'], 
                    row['bairro'], row['cep'], row['uf'], row['municipio'],
                    row['ddd1'], row['telefone1'], row['ddd2'], row['telefone2'],
                    row['ddd_fax'], row['fax'], row['email'], 
                    row['situacao_especial'], row['data_situacao_especial']
                )
                values.append(value)

            # Query de inserção
            insert_query = """
            INSERT INTO estabelecimentos (
                cnpj_basico, cnpj_ordem, cnpj_dv, identificador_matriz_filial,
                nome_fantasia, situacao_cadastral, data_situacao_cadastral,
                motivo_situacao_cadastral, cidade_exterior, pais, data_inicio_atividade,
                cnae_fiscal_principal, cnae_fiscal_secundario, tipo_logradouro,
                logradouro, numero, complemento, bairro, cep, uf, municipio,
                ddd1, telefone1, ddd2, telefone2, ddd_fax, fax, email,
                situacao_especial, data_situacao_especial
            ) VALUES %s
            ON CONFLICT (cnpj_basico, cnpj_ordem, cnpj_dv) DO NOTHING
            """

            try:
                # Inserção em batch
                extras.execute_values(cursor, insert_query, values)
                conn.commit()
                total_linhas_inseridas += len(values)
                print(f"{total_linhas_inseridas} linhas inseridas até agora.")
            except Exception as e:
                print(f"Erro ao inserir dados na tabela estabelecimentos: {e}")
                
                # Log detalhado para a linha problemática
                for val in values:
                    try:
                        # Testa a inserção linha por linha para identificar a exata com erro
                        cursor.execute(insert_query, (val,))
                        conn.commit()
                    except Exception as linha_erro:
                        print(f"Erro específico para o email {val[27]}: {linha_erro}")
                        break  # Sai do loop interno após encontrar o erro

                break  # Sai do loop de chunks após logar o erro

        cursor.close()
        print("Processo concluído. Todos os dados foram inseridos na tabela estabelecimentos.")
        
    except Exception as e:
        print(f"Erro ao inserir dados na tabela estabelecimentos: {e}")

# Caminho do arquivo corrigido
csv_path = "/home/guilherme/web_scrapy/dados_cnpj_extraidos/K3241.K03200Y2.D41012.ESTABELE_CORRIGIDO.csv"

# Conectar ao banco e inserir os dados em chunks
conn = conectar_banco()
if conn:
    inserir_dados_em_chunks(conn, csv_path, chunksize=10)  # Ajuste o chunksize manualmente se necessário
    conn.close()

```

```python
import pandas as pd
import numpy as np
from psycopg2 import sql

# Função para verificar e inserir uma referência ausente em uma tabela auxiliar
def verificar_e_inserir_referencia(cursor, tabela, coluna, valor):
    if valor is not None:  # Ignora valores nulos
        cursor.execute(sql.SQL("SELECT 1 FROM {} WHERE {} = %s").format(
            sql.Identifier(tabela), sql.Identifier(coluna)), [valor])
        if cursor.fetchone() is None:
            cursor.execute(sql.SQL("INSERT INTO {} ({}) VALUES (%s)").format(
                sql.Identifier(tabela), sql.Identifier(coluna)), [valor])
            print(f"Inserido valor ausente '{valor}' na tabela '{tabela}'.")

# Conectar ao banco de dados
conn = conectar_banco()

# Caminho do arquivo e parâmetros de chunking
csv_path = "/home/guilherme/web_scrapy/dados_cnpj_extraidos/K3241.K03200Y2.D41012.ESTABELE"
chunk_size = 1000  # Tamanho de cada chunk
total_inserted = 0

# Ler o arquivo CSV em chunks
try:
    for chunk in pd.read_csv(csv_path, delimiter=";", header=None, encoding="ISO-8859-1", 
                             low_memory=False, dtype=str, chunksize=chunk_size):

        # Renomear as colunas para corresponder à estrutura da tabela
        chunk.columns = [
            "cnpj_raiz", "cnpj_ordem", "cnpj_dv", "identificador_matriz_filial", "nome_fantasia",
            "situacao_cadastral", "data_situacao_cadastral", "motivo_situacao_cadastral",
            "nome_cidade_exterior", "codigo_pais", "data_inicio_atividade", "cnae_fiscal",
            "cnae_fiscal_secundaria", "tipo_logradouro", "logradouro", "numero", "complemento",
            "bairro", "cep", "uf", "codigo_municipio", "ddd1", "telefone1", "ddd2", "telefone2",
            "ddd_fax", "fax", "correio_eletronico", "situacao_especial", "data_situacao_especial"
        ]

        # Convertendo colunas de data para o formato 'YYYY-MM-DD' e mantendo valores nulos como None
        chunk["data_situacao_cadastral"] = pd.to_datetime(chunk["data_situacao_cadastral"], format="%Y%m%d", errors="coerce").dt.strftime('%Y-%m-%d').replace({pd.NaT: None})
        chunk["data_inicio_atividade"] = pd.to_datetime(chunk["data_inicio_atividade"], format="%Y%m%d", errors="coerce").dt.strftime('%Y-%m-%d').replace({pd.NaT: None})
        chunk["data_situacao_especial"] = pd.to_datetime(chunk["data_situacao_especial"], format="%Y%m%d", errors="coerce").dt.strftime('%Y-%m-%d').replace({pd.NaT: None})

        # Substituir valores NaN por None
        chunk = chunk.where(pd.notnull(chunk), None)

        # Processar cada linha no chunk
        with conn.cursor() as cursor:
            for _, row_data in chunk.iterrows():
                row = tuple(row_data)
                
                # Verificar e inserir referências nas tabelas auxiliares, se ausentes
                verificar_e_inserir_referencia(cursor, 'paises', 'codigo', row[9])         # codigo_pais
                verificar_e_inserir_referencia(cursor, 'cnaes', 'codigo', row[11])         # cnae_fiscal
                verificar_e_inserir_referencia(cursor, 'municipios', 'codigo', row[20])    # codigo_municipio
                verificar_e_inserir_referencia(cursor, 'motivos', 'codigo', row[7])        # motivo_situacao_cadastral

                # Inserir na tabela estabelecimentos
                insert_query = sql.SQL("""
                    INSERT INTO estabelecimentos (cnpj_raiz, cnpj_ordem, cnpj_dv, identificador_matriz_filial, nome_fantasia,
                                                 situacao_cadastral, data_situacao_cadastral, motivo_situacao_cadastral,
                                                 nome_cidade_exterior, codigo_pais, data_inicio_atividade, cnae_fiscal,
                                                 cnae_fiscal_secundaria, tipo_logradouro, logradouro, numero, complemento,
                                                 bairro, cep, uf, codigo_municipio, ddd1, telefone1, ddd2, telefone2,
                                                 ddd_fax, fax, correio_eletronico, situacao_especial, data_situacao_especial)
                    VALUES ({})
                """).format(sql.SQL(', ').join(sql.Placeholder() if val is not None else sql.SQL("NULL") for val in row))
                
                cursor.execute(insert_query, [val for val in row if val is not None])
            
            conn.commit()
            total_inserted += len(chunk)
            print(f"{total_inserted} linhas foram inseridas com sucesso.")

except Exception as e:
    print(f"Erro ao inserir dados: {e}")
finally:
    conn.close()
    print(f"Inserção finalizada. Total de linhas inseridas: {total_inserted}")
```
